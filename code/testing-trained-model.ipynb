```python
# To Test Data using trained file

# Step 1: Import Libraries
import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Step 2: Load the Trained Model
model_path = "/kaggle/input/trained-model-final/socofing_fingerprint_model.h5"  # Provide the path to your trained model
model = load_model(model_path)
print("Model loaded successfully!")
```

```python
# Step 3: Function to Predict the Image
def predict_image(model, img_path):
    img = load_img(img_path, target_size=(128, 128))  # Load the image with the same size as the training images
    img_array = img_to_array(img) / 255.0  # Normalize the image to [0, 1] range
    img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension
    prediction = model.predict(img_array)  # Make prediction
    return "Altered" if prediction[0][0] < 0.5 else "Real"
```

```python
# Step 4: Function to Display the Custom Image
def display_image(image_path):
    img = load_img(image_path, target_size=(128, 128))  # Resize the image
    img_array = img_to_array(img) / 255.0  # Normalize to [0, 1] range
    plt.figure(figsize=(6, 6))
    plt.imshow(img_array)
    plt.axis('off')
    plt.title("Custom Test Image")
    plt.show()
```

```python
# Step 5: Provide the Path to Your Custom Image
custom_image_path = "/kaggle/input/ai-generated-images-final/AI generated fingerprints/Gemini 7.bmp"  # Provide the path to your custom image
```

```python
# Step 6: Display the Custom Image
display_image(custom_image_path)
```

```python
# Step 7: Classify the Custom Image
result = predict_image(model, custom_image_path)
print(f"The fingerprint is classified as: {result}")
```
